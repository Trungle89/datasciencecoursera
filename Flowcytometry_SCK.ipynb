```python
from google.colab import files

uploaded = files.upload()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from fcsparser import parse

# Process each uploaded FCS file
for filename in uploaded.keys():
    # Load the FCS file
    meta, data = parse(filename, reformat_meta=True)

    # Convert data to a DataFrame
    df = pd.DataFrame(data)

    # Display basic information
    print(f"Processing file: {filename}")
    print(df.head())

    # Generate and save plots

    # Scatter plot example (adjust the columns to your needs)
    plt.figure(figsize=(10, 8))
    plt.scatter(df.iloc[:, 0], df.iloc[:, 1], alpha=0.5)
    plt.title(f"Scatter Plot: {filename}")
    plt.xlabel(df.columns[0])
    plt.ylabel(df.columns[1])
    plt.savefig(f"{filename}_scatter_plot.png")
    plt.show()
    plt.close()  # Close the figure to prevent memory issues

    # Density plot example
    plt.figure(figsize=(10, 8))
    df.iloc[:, 0].plot(kind='density', alpha=0.5)
    plt.title(f"Density Plot: {filename}")
    plt.xlabel(df.columns[0])
    plt.savefig(f"{filename}_density_plot.png")
    plt.show()
    plt.close()  # Close the figure to prevent memory issues
```
